```python
import matplotlib.pyplot as plt
import pandas
resultados = {"Numero":[], "LenUlam":[]}

for i in range(3,65536):
    resultados["Numero"].append(i)
    Count = 1
    while i > 1:
        if i % 2 == 0:
            Count += 1
            i = i / 2
        else:
            Count += 1
            i = 3 * i + 1
    resultados["LenUlam"].append(Count)
```

### Tempo de execução de 1.6s

```python
print("Tamanho médio da sequência:",sum(resultados["LenUlam"])/len(resultados["LenUlam"]))
```

```python
df = pandas.DataFrame(resultados)
df.plot(x="Numero",y="LenUlam")
plt.show()
```
